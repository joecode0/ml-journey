```python
import pandas as pd
data_folder = '../data/'

df = pd.read_csv(data_folder + 'house-prices/train.csv')
df.head()
```

```python
import sys
# For every column, print the name, type, top 3 values with their counts, and the percentage of missing values
for col in df.columns:
    nunique = df[col].nunique()
    missing = df[col].isnull().sum() / len(df)
    val_counts =  df[col].value_counts()
    if len(val_counts) > 3:
        try:
            val_counts = val_counts.iloc[:3]
        except:
            print("Error with column {}".format(col))
            print(val_counts)
            sys.exit(0)
    print("{} unique, {}% missing".format(nunique, missing))
    print(val_counts)
    print("-------------------------------------------------")
```
